```python
import requests
from fake_useragent import UserAgent
```

```python
ua = UserAgent(verify_ssl=False)
USER_AGENT = ua.random
```

```python
url = 'https://www.smzdm.com/p/24341561/#comments'
```

```python
headers = {'user-agent': USER_AGENT}
resp = requests.get(url, headers=headers)
```

```python
resp.text
```

```python
from scrapy.selector import Selector 
```

```python
selector = Selector(resp)
tags = selector.xpath('//div[@class="comment_conBox"]')
```

```python
tags[0].xpath('./div/div[@class="time"]/meta/@content').get()
```

```python
tags[0].xpath('./div/div/p/span[@itemprop="description"]/text()').get()
```

```python
times = []
shorts = []
for tag in tags:
    time = tag.xpath('./div/div[@class="time"]/meta/@content').get()
    short = tag.xpath('./div/div/p/span/text()').get()
    times.append(time)
    shorts.append(short)
```
