```python
import pandas as pd
import numpy as np
```

```python
group = ['x', 'y', 'z']
df = pd.DataFrame({
    "group":[group[x] for x in np.random.randint(0,len(group),10)] ,
    "age":np.random.randint(15,50,10),
    "salary":np.random.randint(5,50,10),
    })
df
```

```python
# 导出为.xlsx文件
df.to_excel(excel_writer=r'excel.xlsx')
```

```python
# 设置Sheet名称
df.to_excel(excel_writer=r'excel1.xlsx', sheet_name='sheet_2')
```

```python
# 设置索引,设置参数index=False就可以在导出时把这种索引去掉
# 设置要导出的列
df.to_excel(excel_writer=r'excel2.xlsx', index=False)
```

```python
df = df.append({'group': np.nan,
          'age': 40,
          'salary': 67}, ignore_index=True)
df
```

```python
# 设置要导出的列
df.to_excel(excel_writer=r'excel3.xlsx', 
            index=False, 
            columns=['group', 'age'], 
            encoding='utf-8', 
            na_rep=0,
            inf_rep=0) # columns=['group', 'age']无效????
```

```python
# 导出为.csv文件
df.to_csv(r'csv.csv', encoding='utf-8')
```
