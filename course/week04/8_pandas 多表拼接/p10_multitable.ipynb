### 注意：合并后对组合关系！

```python
import pandas as pd
import numpy as np
```

```python
group = ['x', 'y', 'z']
data1 = pd.DataFrame({
    'group': [group[x] for x in np.random.randint(0, len(group), 10)],
    'age': np.random.randint(15, 50, 10)
})

data2 = pd.DataFrame({
    "group":[group[x] for x in np.random.randint(0,len(group),10)] ,
    "salary":np.random.randint(5,50,10),
    })

data3 = pd.DataFrame({
    "group":[group[x] for x in np.random.randint(0,len(group),10)] ,
    "age":np.random.randint(15,50,10),
    "salary":np.random.randint(5,50,10),
    })
```

```python
data1.sort_values('group')
```

```python
data2.sort_values(['group', 'salary'])
```

```python
data3.sort_values(['group', 'age'])
```

```python
# 一对一组合
pd.merge(data1, data2)
```

```python
# 多对一组合
pd.merge(data3, data2, on='group').sort_values(['group', 'age'])
```

```python
#多对多,group、salary相同的才合并。
pd.merge(data3, data2)
```

```python
# 连接键类型，解决没有公共列问题,即指定公共列，具体来说：data3的age==data2的salary的数据进行合并。
pd.merge(data3, data2, left_on='age', right_on='salary')
```

### 连接方式
#### 内连接，不指明连接方式，默认都是内连接

```python
pd.merge(data3, data2, on='group', how='inner')
```

```python
# 左连接 left
# 右连接 right
# 外连接 outer
```

```python
# 纵向拼接
pd.concat([data1, data2])
```

```python
pd.concat([data2, data3])
```
