### GOD

```python
#默认继承object
class Human:
    # 静态字段
    live = True
    
    def __init__(my, name):
        # 普通字段
        my.name = name
```

```python
man = Human('Adam')
woman = Human('Eve')
```

```python
man.__class__
```

```python
woman.__class__
```

```python
Human.__class__
```

```python
man.name
```

### 查看属性与方法

```python
Human.__dict__
```

```python
man.__dict__
```

```python
woman.__dict__
```

#### 实例可以使用普通字段也可以使用静态字段

```python
man.live
```

```python
man.live = False
man.__dict__
```

```python
Human.live
```

```python
woman.live
```

#### 类可以使用静态字段,也可以为类添加静态字段

```python
Human.newattr = 1
```

```python
dir(Human)
```

```python
dir(man)
```

```python
Human.__dict__
```

### 增加属性和方法

```python
setattr(Human, 'dead', False)
```

```python
Human.__dict__
```

```python
setattr(man, 'income', 9999)
man.__dict__
```

```python
dir(man)
```

```python
# 内置类型不能增加属性和方法
setattr(list, 'newattr', 4)
```
